```python
import enerhabitat as eh
import pandas as pd
import matplotlib.pyplot as plt
from enerhabitat import config
```

```python
print(sorted(config.materials.keys()))
```

```python
epw_file = "epw/MEX_MOR_Cuernavaca-Matamoros.Intl.AP.767260_TMYx.2009-2023.epw"

wall = eh.System(eh.Location(epw_file))
wall.azimuth = 90
wall.tilt = 90

wall.absortance = 1
wall.layers = [("mortero", 0.02), ("tabique", 0.14)]
wall.location.meanDay(month=5, year=2026)
wall.Tsa()

blanco = wall.solve()

blanco = pd.concat([blanco, wall.Tsa().asfreq("10min")], axis=1)

blanco
```
